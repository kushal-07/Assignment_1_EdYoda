```python
# 1Q) Write a Python program to get the Fibonacci series between 0 to 50


def fibonacci(n):
    a = 0
    b = 1
    if n < 0:
        print("Incorrect input")
    elif n == 0:
        return a
    elif n == 1:
        return b
    else:
        print(1)
        for i in range(2,n+1):
            c = a + b
            print(c)
            a = b
            b = c
        
fibonacci(9)
```

```python
# Q3 Write a Python program to count the number of even and odd numbers from a series of numbers.

n = int(input("Enter number of elements : "))
lst=[] 

for i in range(0, n):
    ele = int(input())
 
    lst.append(ele) 
even = 0
odd = 0 
for x in lst:
    if x%2 ==0:
        even = even +1
    else: 
        odd = odd+1
        
print("No of even numbers in the given list are", even)
print("No of off numbers in the given list are", odd)
```
